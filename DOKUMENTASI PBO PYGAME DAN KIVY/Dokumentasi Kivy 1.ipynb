# __KIVY__

__A. Pengertian Kivy__

   Kivy ini sendiri merupakan framework yang dibangun menggunakan library dari bahasa pemrograman Python yang bersifat Open Source. Tujuan dikembangkannya framework ini agar dapat membantu developer secara cepat dalam mengembangkan aplikasi yang memiliki tampilan antarmuka inovatif seperti aplikasi yang mendukung multitouch.
   
   Framework ini sendiri dapat dijalankan pada sistem operasi Windows, Mac OS X, Linux, Android, hingga Raspberry Pi. Selain itu, Kivy mendukung untuk input beragam perangkat seperti WM_Touch, WM_Pen, Mac OS X Trackpad, Magi Mouse, Mtdev, Linux Kernel HID, dan TUIO.

__B. Cara Instalasi Kivy__

Buat / buka proyek Anda di PyCharm
    
   Itu biasanya yang diperlukan. Jika Anda memiliki beberapa penerjemah Python, Anda mungkin perlu memilih salah satu yang Anda instal Kivy.

1. File -> Pengaturan -> Proyek -> Project Interpreter
2. Klik simbol tambah di sebelah kanan  proyek dan pilih penerjemah yang sesuai

# __C. Penjelasan Class__

__a. class HourLabel__

Pada Class HourLabel terdapat 2 method yaitu :

*	Pos_offset untuk Kembalikan vektor yang menggambarkan di mana label harus diposisikan, relatif terhadap pusat jam, menurut jamnya atribut.
*	Update untuk Perbarui teks dan posisi label agar sesuai dengan atributnya.

__b. class DigitalTime__

Pada class DigitalTime digunakan untuk tampilan jam digital yang mempunyai 1 method yaitu method update.

__c. class DigitalTime12__

Pada class ini akan menampilkan jam digital 24 jam dengan detik. Yang mempunyai 2 method yaitu fmt dan update, method update digunakan untuk memperbarui tampilan ke waktu baru.

__d. class DigitalTime12__

Pada class ini digunakan untuk menampilkan jam digital 12 jam tanpa detik. Yang mempunyai method yang sama pada class digitaltime24.

__e. class ClockHand__

Pada class ini akan menampilkan kelas dasar untuk jarum jam. kelas ini terdapat 3 method yaitu:

* Points untuk menghitung point yang dibutuhkan untuk menggambar jarum jam pada sudutnya saat ini.
* Update untuk memperbarui jarum jam ke nilai yang ditentukan.
* Construct digunakan buat instruksi untuk menggambarkan jarum jam.

__f. class HourHand__

Class ini merupakan jarum jam 

__g. class MinuteHand__

Class ini merupakan jarum menit


__h. class SecondHand__

Class ini merupakan jarum detik

__i. Class ClockFace__

Pada class ini akan menampilkan jam analog yang mempunyai 5 method yaitu:

* Update yang digunakan untuk memperbarui semua jarum jam ke nilai yang ditentukan.
* Update_hour_labels untuk memanggil metode pembaruan pada semua label jam.
* Flip_next_hour_label yaitu panggilan balik. Membalik label berikutnya secara beruntun.
• Flip_hour_labels memulai urutan membalik label jam dari desimal ke nomor Romawi, atau sebaliknya.
• Start untuk Inisialisasi dan memulai jam analog

__j. Class WorldClock__

Class ini menampilkan jam waktu dalam kata-kata bahasa inggris. Class ini terdapat method update yang berguna untuk memperbarui tampilan jam kata dengan jam dan menit yang ditentukan nilai-nilai.

__k. Class DateDisplay__

Class ini akan menampilkan tanggal saat ini dalam bentuk panjang daengan format dd/mm/yy. Class ini memiliki method update, methode ini memperbarui tampilan tanggal ke y/m/d yang ditentukan. Untuk menghindari ketidakeffisien pembaruan akan dilewati jika tanggal tidak berubah sejak penggilan terakhir. Argumen 'force' ada untuk menggantikan ini jika perhitungan ulang diperlukan.

__l. Class BKClock__

Class ini akan menampilkan layout dari class :
* clock_face
* digital_12
* digital_24
* word_clock

class ini juga mempunyai method yaitu method update yang digunakan untuk panggilan balik tunggal untuk mendapatkan waktu saat ini dan memasukkannya ke tampilan jam.

__m. Class BKClockApp__

Class ini mempunyai method build yang untuk nilai kembali class BKClock.
